```python
#cheaking for outliers
dataset.boxplot()
```

```python
rm_features = [
    "MDVP:Fo(Hz)", "MDVP:Fhi(Hz)", "MDVP:Flo(Hz)", "NHR", "HNR", "DFA"
]
```

```python
feature_dataset = dataset.drop(rm_features, axis = 1)
```

```python
feature_dataset.head()
```

```python
feature_dataset.shape
```

# Data visualization

```python
feature_dataset.hist(figsize=(14,14))
plt.show()
```

```python
sns.pairplot(feature_dataset,hue='status')
```

Tranfroming the skewed data into normal distribution


Data is left Skewed applying log Transformation

```python
feature_dataset["MDVP:Jitter(%)"] = feature_dataset["MDVP:Jitter(%)"].transform(np.log)
```

```python
feature_dataset["MDVP:Jitter(%)"].plot.hist()
```

```python
feature_dataset["MDVP:Jitter(Abs)"] = feature_dataset["MDVP:Jitter(Abs)"].transform(np.log)
feature_dataset["MDVP:Jitter(Abs)"].plot.hist()
```

```python
feature_dataset["MDVP:RAP"]=feature_dataset["MDVP:RAP"].transform(np.log)
feature_dataset["MDVP:PPQ"]=feature_dataset["MDVP:PPQ"].transform(np.log)
feature_dataset["Jitter:DDP"] = feature_dataset["Jitter:DDP"].transform(np.log)
feature_dataset["MDVP:Shimmer"] = feature_dataset["MDVP:Shimmer"].transform(np.log)
feature_dataset["MDVP:Shimmer(dB)"] = feature_dataset["MDVP:Shimmer(dB)"].transform(np.log)
feature_dataset["Shimmer:APQ3"] = feature_dataset["Shimmer:APQ3"].transform(np.log)
feature_dataset["Shimmer:APQ5"] = feature_dataset["Shimmer:APQ5"].transform(np.log)
feature_dataset["Shimmer:DDA"] = feature_dataset["Shimmer:DDA"].transform(np.log)
feature_dataset["MDVP:APQ"] = feature_dataset["MDVP:APQ"].transform(np.log)
```

```python
feature_dataset.head()
```

```python
#Saving the transformed dataset
feature_dataset.to_csv("transformed_data.csv")
```
